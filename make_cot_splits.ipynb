```python
import pandas as pd
import numpy as np

cot_examples_df = pd.read_csv('data/candidate_cot_dataset.csv')
```

```python
# shuffle dataframe
cot_examples_df = cot_examples_df.sample(frac=1, random_state=1).reset_index(drop=True)
print("df size: ", len(cot_examples_df))
```

```python
for example in cot_examples_df["id"]:
     print(example)
```

```python
# make splits
cot_examples_subsets = []
split_size = 5
for i in range(1 ,len(cot_examples_df) + 1): # start at to avoid 0 % split_size
    if i % split_size == 0:
        cot_examples_subsets.append(cot_examples_df.iloc[i-split_size:i])
```

```python
for i, subset in enumerate(cot_examples_subsets, start=1):
    save_path = f'data/cot_ic_set_{i}.csv'
    print("saving at: ", save_path)
    subset.to_csv(save_path, index=False)
```
